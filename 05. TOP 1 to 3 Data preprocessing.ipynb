# 2. 데이터 전처리

```python
df_payment = payment[['발급회원번호', '이용금액_쇼핑','이용금액_요식','이용금액_교통','이용금액_의료','이용금액_납부','이용금액_교육',
'이용금액_여유생활','이용금액_사교활동','이용금액_일상생활','이용금액_해외']]
```

```python
df_payment['전체금액'] = df_payment.iloc[:,1:].sum(axis=1)
```

```python
df_pay = df_payment[~(df_payment['전체금액'] == 0)].drop('전체금액',axis=1).reset_index(drop=True)
```

```python
df_pay.head()
```

```python
df_pay.iloc[:,1:].idxmax(axis=1)
```

# 3. TOP3 카테고리 추출

```python
df_pay['category'] = df_pay.iloc[:,1:].apply(lambda s: s.abs().nlargest(3).index.tolist(), axis=1)
```

```python
df_pay['category']
```

```python
df_pay.to_csv('top_payment.csv',index=False)
```

```python
df_pay.explode('category')
```

```python
df_top = pd.DataFrame(df_pay['category'].to_list(),columns=['top1','top2','top3'])
df_top.to_csv('top_category.csv',index=False)
```

```python
df_top['top1'] = df_top['top1'].str.replace('이용금액_','')
df_top['top2'] = df_top['top2'].str.replace('이용금액_','')
df_top['top3'] = df_top['top3'].str.replace('이용금액_','')
```

```python
df_top = pd.concat([df_top, df_pay['발급회원번호']], axis=1)
```

```python
df_top.to_csv('top_category.csv',index=False)
```

# 4. 고객 개인정보와 TOP 결제 카테고리 합치기

```python
cust_info
df_all = pd.merge(cust_info, df_top, how='right', on = '발급회원번호')
```
